```python
def list_json_files(directory_path):
    json_files = [file for file in os.listdir(directory_path)
                  if file.lower().endswith(".json") and os.path.isfile(os.path.join(directory_path, file))]
    return json_files
```

Logistical movement for downstream processing

```python
import shutil
```

```python
# For each .json file created from parse()
for json_file in list_json_files('../landing-ai-sandbox-data/intermediary/'):
    # Open and load the JSON file
    with open(f'../landing-ai-sandbox-data/intermediary/{json_file}', 'r') as f:
        chunked_json_file = json.load(f)
        
    # json_file are named with document_num_startTime_endTime.json to retreive first part
    document_num = '_'.join(json_file.split('_')[:2])
    
    # Make directory to store .md and .pdf file
    os.makedirs(f"../landing-ai-sandbox-data/input_data/{document_num}", exist_ok=True)
    
    # Write out the .md files
    with open(f"../landing-ai-sandbox-data/input_data/{document_num}/{document_num}.md", "w") as file:
        # Retrieve the .md part and write out
        file.write(chunked_json_file['markdown'])
    
    # Copy the original .pdf file to folder
    shutil.copy2(f'{base_dir}/{document_num}.pdf', os.path.join(f"../landing-ai-sandbox-data/input_data/{document_num}/", f"{document_num}.pdf"))
```
